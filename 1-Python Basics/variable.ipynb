# Dynamic typing

```python
# Python allows the type of a variable to change as the program executes

var = 10 #int
print(var, type(var))

var = "Hello" #str
print(var, type(var))

var = 10.22 #float
print(var, type(var))
```

# Simple Calculator

```python
num1 = float(input("Enter first number : "))
num2 = float(input("Enter second number : "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("First Num : ", num1, "\nSecond Num : ", num2 )
print("Sum : ", sum)
print("Difference : ", difference)
print("Product : ", product)
print("Quotient : ", quotient)
```
