# Operators

## Arithmetic Operators

```python
a = 10
b = 5

add_result = a+b
sub_result = a-b
mul_result = a*b
div_result = a/b
float_div_result = a//b
modulus_result = a%b
exponent_result = a**b

print(add_result)
print(sub_result)
print(mul_result)
print(div_result)
print(float_div_result) 
print(modulus_result) 
print(exponent_result) 
```

## Comparision Operator
