#### If statement :

```python
age = 24

if age >= 18:
    print("You are a adult.")
```

#### Else :

```python
age = 12

if age >= 18:
    print("Eligible for vote")
else:
    print("Not eligible for vote")
```

```python
age = 10

if age < 12:
    print("You are minor")
elif age<18:
    print("You are teenager")
else:
    print("You are adult")
```

#### Nested if

```python
num  = int(input("Enter a number"))

if num > 0:
    print("Positive")
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
else:
    print("Negetive")
    
```

### Practical Example

#### Determine if a year is leap year or not using neasted if.

```python
year = int(input("Enter the year"))

if year % 4 == 0:
    print("Leap")
else:
    print("Not leap")
```
